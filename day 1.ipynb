```python
import pandas as pd
df1={
    "name":['aaa','bbb','ccc','ddd'],
    'age':[10,20,30,40],
    'salary':[1000,2000,3000,8000]
}
df2={
     "name":['uuu','bbb','vvv','ddd'],
    'age':[10,20,30,40],
    'salary':[1000,5000,3000,8000]
}
d1=pd.DataFrame(df1)
d2=pd.DataFrame(df2)
print(d1)
print(d2)
c=pd.concat([d1,d2],axis=1)
print(c)
m=pd.merge(d1,d2, on=['age'])
print(m)
```

```python
 #groupe
data1={
    "department":["HR",'DATA','HR','IT','IT','HR'],
    'name':['aaa','bbb','aaa','ddd','eee','fff'],
    'salary':[10,60,30,100,55,25],
    'bonus':[10000,55000,25000,30000,45000,35000],
}
df=pd.DataFrame(data1)
print(df)
d=df.groupby(['department','name'])['salary'].sum()
print(d)
c=df.groupby('department')[['salary','bonus']].sum()
print(c)
f=df[df['salary']>50]
print(f)
g=df.sort_values(['salary'],ascending=False)
print(g)
```

```python
import pandas as pd

data = {
    'name': ['aaa', 'bbb', 'aaa', 'bbb'],
    'subject': ['math', 'math', 'eng', 'eng'],
    'marks': [90, 80, 85, 70]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Pivot the table
pivot_df = df.pivot(index='name', columns='subject', values='marks')

print("\nPivoted DataFrame:\n", pivot_df)
p=df.info()
print(p)
```
